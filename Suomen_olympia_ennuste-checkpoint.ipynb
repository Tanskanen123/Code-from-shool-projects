```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Otetaan data
df = pd.read_csv("archive/athlete_events.csv", sep=",")

#Luodaan kesa taulukko, jossa on suomen kilpailijat kesäolympialaisista
kesa = df.loc[(df['NOC'] == 'FIN') & (df['Season'] == 'Summer')]
#Suodatetetaan häviäjät, eli NA-arvot, eli Null-arvot
kesa = kesa[kesa.Medal.notnull()]

#Luodaan talvi taulukko, jossa on suomen kilpailijat talviolympialaisista
talvi = df.loc[(df['NOC'] == 'FIN') & (df['Season'] == 'Winter')]
#Suodatetetaan häviäjät, eli NA-arvot, eli Null-arvot
talvi = talvi[talvi.Medal.notnull()]

#Luodaan talvi ja kesa pöydistä uudet pöydät, joissa on vuosi, ja kyseisenä vuonna voitetut mitalit
talvi_tulos = talvi.pivot_table(index = ['Year'], aggfunc ='size')
kesa_tulos = kesa.pivot_table(index = ['Year'], aggfunc ='size')


#Luodaan eirlliset taulut kesien ja talvien vuosille ja mitalien määrille

#Talvi
talvi_tulos_mitalit = []
talvi_tulos_vuodet = [1924,1928,1932,1936,1948,1952,1956,1960,1964,1968,1972,1976,1980,1984,1988,1992,1994,1998,2002,2006,2010,2014]

for x in talvi_tulos:
    talvi_tulos_mitalit.append(x)

    
#Kesä
kesa_tulos_mitalit = []
kesa_tulos_vuodet = [1906,1908,1912,1920,1924,1928,1932,1936,1948,1952,1956,1960,1964,1968,1972,1976,1980,1984,1988,1992,1996,2000,2004,2008,2012,2016]

for x in kesa_tulos:
    kesa_tulos_mitalit.append(x)
    
    
    
#Luodaan ennuste talvelle
#Jos haluat ennusteen kesäälle, niin poista "talvi" ja kirjoita tilalle "kesa"
from scipy import stats

slope, intercept, r, p, std_err = stats.linregress(talvi_tulos_vuodet, talvi_tulos_mitalit)

#Ennuste
def ennuste(x):
  return slope * x + intercept


#Ennustetaan
tulevat_mitalit = ennuste(2018)
print(tulevat_mitalit)
tulevat_mitalit = ennuste(2022)
print(tulevat_mitalit)

mymodel = list(map(ennuste, talvi_tulos_vuodet))

plt.scatter(talvi_tulos_vuodet, talvi_tulos_mitalit)
plt.plot(talvi_tulos_vuodet, mymodel)
print(r)
plt.show() 
```
